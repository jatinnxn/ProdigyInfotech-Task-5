# Task 5

Analyze traffic accident data to identify patterns related to road conditions,weather,and time of day. Visualize accident hotspots and contributing factors.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = 'RTA Dataset.csv'
data = pd.read_csv(file_path)

# 1. Data Cleaning
# Remove duplicates
data = data.drop_duplicates()

# Convert 'Time' column to datetime to extract hour
data['Time'] = pd.to_datetime(data['Time'], errors='coerce')
data['Hour'] = data['Time'].dt.hour

# Drop rows with missing 'Time' values as they are essential for analysis
data = data.dropna(subset=['Time'])

# 2. Exploratory Data Analysis (EDA)
# Summary statistics for numerical columns
numerical_summary = data.describe()

# Summary of unique values for categorical columns
categorical_summary = data.describe(include='object')

# Print out the summaries for review
print("Numerical Summary:\n", numerical_summary)
print("\nCategorical Summary:\n", categorical_summary)

# 3. Handling Missing Values
# Drop rows with missing values in essential columns (for our analysis)
essential_cols = ['Road_surface_conditions', 'Weather_conditions', 'Accident_severity', 'Hour']
data_cleaned = data.dropna(subset=essential_cols)

# 4. Data Visualization
# Set up subplots for visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Traffic Accident Patterns by Conditions and Time')

# Accident distribution by road surface conditions
sns.countplot(data=data_cleaned, x='Road_surface_conditions', hue='Accident_severity', ax=axes[0, 0])
axes[0, 0].set_title('Accidents by Road Surface Conditions')
axes[0, 0].set_xlabel('Road Surface Condition')
axes[0, 0].set_ylabel('Accident Count')
axes[0, 0].tick_params(axis='x', rotation=45)

# Accident distribution by weather conditions
sns.countplot(data=data_cleaned, x='Weather_conditions', hue='Accident_severity', ax=axes[0, 1])
axes[0, 1].set_title('Accidents by Weather Conditions')
axes[0, 1].set_xlabel('Weather Condition')
axes[0, 1].set_ylabel('Accident Count')
axes[0, 1].tick_params(axis='x', rotation=45)

# Accident distribution by hour of the day
sns.histplot(data=data_cleaned, x='Hour', hue='Accident_severity', multiple='stack', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Accidents by Time of Day')
axes[1, 0].set_xlabel('Hour of the Day')
axes[1, 0].set_ylabel('Accident Count')

# Accident hotspots by day of the week
sns.countplot(data=data_cleaned, x='Day_of_week', hue='Accident_severity', ax=axes[1, 1], order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
axes[1, 1].set_title('Accidents by Day of the Week')
axes[1, 1].set_xlabel('Day of the Week')
axes[1, 1].set_ylabel('Accident Count')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
